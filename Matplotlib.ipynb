# Line Chart

```
import matplotlib.pyplot as plt
import numpy as np
```

```
x = np.linspace(0,10,100)
```

```
y = np.sin(x)
```

```
plt.plot(x,y)
```

```
plt.xlabel("time")
plt.ylabel("some time method")
plt.title("sin plot")
plt.plot(x,y)
```

```
!wget https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/linear_regression_class/data_1d.csv
```

```
import pandas as pd
A = pd.read_csv('data_1d.csv',header=None).values
```

```
x = A[:,0]
y = A[:,1]
```

```
plt.scatter(x,y)
```

```
x_line = np.linspace(0,100,100)
y_line = 2*x_line+1
```

```
plt.scatter(x,y)
plt.plot(x_line,y_line)
```
