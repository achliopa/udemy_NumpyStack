```
# we use a script from TF2 course to upload acsv into colabs
!wget https://raw.githubusercontent.com/lazyprogrammer/machine_learning_examples/master/linear_regression_class/data_2d.csv
```

```
# we see a script to manual load the csv into a numpy array

# NOTE: in class, we assumed the current working directory
#       was linear_regression_class
#       in this file, we assume you are running the script
#       from the directory this file is in

import numpy as np

X = []

for line in open('data_2d.csv'):
  row = line.split(',')
  sample = map(float, row)
  X.append(sample)

X = np.array(X)
print(X)

```

# The apply() Function

```
df["passenges_squared"] = df.apply(lambda row: row["passengers"]*row["passengers"],axis=1)
```

```
df.head(1)
```

```
from datetime import datetime
```

```
datetime.strptime("1949-01","%Y-%m")
```

```
df["dt"] = df.apply(lambda row: datetime.strptime(row["month"],"%Y-%m"),axis=1)
df.head()
```
